```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/kaggle/input/global-ai-job-market-and-salary-trends-2025/ai_job_dataset.csv")
df.dropna(subset=['required_skills'], inplace=True)
all_skills = []

for index, row in df.iterrows():
    skills_list = row['required_skills'].split(',')
    cleaned_skills = [skill.strip() for skill in skills_list]
    all_skills.extend(cleaned_skills)


skills_series = pd.Series(all_skills)
top_15_skills = skills_series.value_counts()
base_color = "#e67cb9"
custom_palette = sns.light_palette(base_color, n_colors=24, reverse=True)

plt.figure(figsize=(12, 9))
ax = sns.barplot(
    x=top_15_skills.values,
    y=top_15_skills.index,
    palette=custom_palette,
    hue=top_15_skills.index,
    legend=False
)

ax.set_title('Top 15 Competenze pi√π Richieste', fontsize=16)
ax.set_xlabel('Numero di Menzioni', fontsize=12)
ax.set_ylabel('Competenze', fontsize=12)
plt.tight_layout()
plt.savefig("competenze.png")
plt.show()
```
