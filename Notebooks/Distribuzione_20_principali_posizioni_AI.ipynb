```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/global-ai-job-market-and-salary-trends-2025/ai_job_dataset.csv")
fig, ax = plt.subplots(figsize=(12, 8))
df['job_title'].value_counts().plot.barh(color="#e67cb9")

ax.set_title('Distribuzione delle 20 Principali Posizioni Lavorative in AI', fontsize=16)
ax.set_xlabel('Numero di Offerte', fontsize=16)

plt.savefig("Distribuzione_20_principali_posizioni_AI.png")
plt.show()
```
