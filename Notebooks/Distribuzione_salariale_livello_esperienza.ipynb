```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/kaggle/input/salaries-for-data-science-jobs/salaries.csv')

sns.set_style("whitegrid")
plt.figure(figsize=(12, 8))

experience_order = ['EN', 'MI', 'SE', 'EX']
labels = ['Entry-Level', 'Mid-Level', 'Senior', 'Executive']
custom_palette = sns.light_palette("#e67cb9", n_colors=4)

sns.boxplot(
    x='experience_level',
    y='salary_in_usd',
    data=df,
    order=experience_order,
    palette=custom_palette
)

plt.title('Distribuzione Salariale per Livello di Esperienza', fontsize=16)
plt.xlabel('Livello di Esperienza', fontsize=12)
plt.ylabel('Stipendio Annuo Lordo (USD)', fontsize=12)
plt.xticks(ticks=range(len(labels)), labels=labels)

plt.gca().get_yaxis().set_major_formatter(
    plt.FuncFormatter(lambda y, p: f'${int(y):,}')
)

plt.tight_layout()
plt.savefig("Distribuzione_salariale_livello_esperienza.png")
plt.show()
```
