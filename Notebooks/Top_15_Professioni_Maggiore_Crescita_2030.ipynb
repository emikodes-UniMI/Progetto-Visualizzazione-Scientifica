```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/kaggle/input/ai-impact-on-job-market-20242030/ai_job_trends_dataset.csv")
top_growth_jobs = df.sort_values('Projected Openings (2030)', ascending=True).tail(15)

sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(12, 10))

for i, job in top_growth_jobs.iterrows():
    y_pos = job['Job Title']
    start_val = job['Job Openings (2024)']
    end_val = job['Projected Openings (2030)']
    growth = end_val - start_val

    ax.plot([start_val, end_val], [y_pos, y_pos], color="#e67cb9", alpha=0.4, linewidth=3, zorder=1)
    
    ax.scatter(start_val, y_pos, color='grey', s=80, zorder=2)
    ax.scatter(end_val, y_pos, color='#e67cb9', s=80, zorder=2)
    
    ax.text(end_val + 200, y_pos, f'+{growth:,.0f}', color='#e67cb9', va='center', ha='left', fontsize=10, fontweight='bold')

ax.set_title('Le 15 Professioni con la Maggiore Crescita Prevista (2024-2030)', fontsize=18, pad=20)
ax.set_xlabel('Numero di Posizioni Aperte', fontsize=14)
ax.set_ylabel('')

ax.spines[['top', 'right', 'left']].set_visible(False)
ax.tick_params(axis='y', length=0)

ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.tight_layout()
plt.savefig("crescita_lavori.png")
plt.show()
```
