```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/kaggle/input/salaries-for-data-science-jobs/salaries.csv')
sns.set_style("whitegrid")
plt.figure(figsize=(10, 7))

size_order = ['S', 'M', 'L']
labels = ['Piccola (<50)', 'Media (50-250)', 'Grande (>250)']

sns.boxplot(
    x='company_size',
    y='salary_in_usd',
    data=df,
    order=size_order,
    palette='coolwarm'
)

plt.title('Distribuzione degli Stipendi per Dimensione Aziendale', fontsize=16)
plt.xlabel('Dimensione Azienda', fontsize=12)
plt.ylabel('Stipendio in USD', fontsize=12)
plt.xticks(ticks=range(len(labels)), labels=labels)

plt.gca().get_yaxis().set_major_formatter(
    plt.FuncFormatter(lambda y, p: f'${int(y):,}')
)

plt.savefig("Distribuzione degli Stipendi per Dimensione Aziendale.png")
plt.tight_layout()
plt.show()
```
