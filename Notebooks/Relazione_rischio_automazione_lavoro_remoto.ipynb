```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/kaggle/input/ai-impact-on-job-market-20242030/ai_job_trends_dataset.csv")
df.dropna(subset=['Job Title', 'Remote Work Ratio (%)', 'Automation Risk (%)'], inplace=True)

job_agg_data = df.groupby('Job Title')[['Remote Work Ratio (%)', 'Automation Risk (%)']].mean().reset_index()

plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(12, 8))

ax = sns.scatterplot(
    data=job_agg_data,
    x='Remote Work Ratio (%)',
    y='Automation Risk (%)',
    color='#e67cb9',
    alpha=0.7,
    s=60  
)

ax.set_title('Relazione tra Lavoro Remoto e Rischio di Automazione', fontsize=18, pad=20)
ax.set_xlabel('Percentuale di Lavoro Remoto Possibile (%)', fontsize=14)
ax.set_ylabel('Rischio di Automazione (%)', fontsize=14)

plt.tight_layout()
plt.savefig("Relazione_rischio_automazione_lavoro_remoto.png")
plt.show()
```
